```python
#Read csv file without a header
df = pd.read_csv('build/output.csv', header=None)

#First column is time
time = df.iloc[:,0]

# second and third are x and y, plot them on x and y axis
plt.figure(figsize=(10,5))
plt.plot(df.iloc[:,1], df.iloc[:,2])
plt.ylabel('Y')
plt.xlabel('X')
plt.title('Position')
plt.grid(True)
plt.show()

#Loop through remaining columns in dataframe
for i in range(1, df.shape[1]):
    plt.figure(figsize=(10,5))
    #Data plot with time as x-axis and each column as y-axis
    plt.plot(time, df.iloc[:,i])
    plt.ylabel('Value')
    plt.xlabel('Time')
    plt.title(f'Column {i}')
    plt.grid(True)
    plt.show()
```
